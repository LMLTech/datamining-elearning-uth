3.1 List (danh sách)
 #Mô tả ngắn
+ List là danh sách có thứ tự thay đổi được (thêm, sửa, xoá).
+ Các phần tử có thể là bất kỳ kiểu dữ liệu nào.
 #Ngữ cảnh áp dụng
- Dùng khi cần quản lý dữ liệu dạng danh sách:
+ Danh sách điểm
+ Danh sách sản phẩm
+ Danh sách sinh viên
+ Dữ liệu cần thêm/xoá liên tục

```python
#ví dụ minh họa code về danh sách điểm
# Tạo danh sách điểm
scores = [7.5, 8.0, 9.2]

# Thêm điểm mới
scores.append(8.8)

# Sửa điểm đầu tiên 7.5
scores[0] = 8.0

# xóa điểm 9.2
del scores[2]

# Duyệt danh sách điểm
print("Danh sách điểm:")
for s in scores:
    print(s)
```

3.3 set - tập hợp không trùng lặp
#Mô tả ngắn
+ Set là tập hợp không có thứ tự, không chứa phần tử trùng nhau.
+ Tự động loại bỏ trùng lặp.
#Ngữ cảnh áp dụng
+ Loại bỏ trùng lặp trong danh sách
+ Kiểm tra phần tử tồn tại
+ Làm các phép toán tập hợp (giao, hợp, hiệu)

```python
# Ví dụ code loại trùng số và tính giao - hợp - hiệu
# Tự động loại trùng
numbers = {1, 2, 2, 3, 4}
print("Set sau khi loại trùng:", numbers)

# Thêm phần tử mới
numbers.add(5)

# Phép toán tập hợp
a = {1, 2, 3}
b = {3, 4, 5}

print("Giao:", a & b)
print("Hợp:", a | b)
print("Hiệu:", a - b)
```

4.Cấu trúc lệnh trong python

```python
# 4.a. Lệnh điều kiện IF / ELIF / ELSE 

# Nhập điểm từ bàn phím (kiểu float vì có điểm thập phân)
score = float(input("Nhập điểm của bạn: "))

# Kiểm tra điểm và phân loại theo từng mức
if score >= 8.5:
    print("Xếp loại: Giỏi")
elif score >= 7:
    print("Xếp loại: Khá")
elif score >= 5:
    print("Xếp loại: Trung bình")
else:
    print("Xếp loại: Yếu")
```

```python
# 4.b. Lệnh lặp FOR
# Người dùng nhập số lần lặp, chương trình in ra từng lần.

# Nhập số lần lặp
n = int(input("Nhập số lần lặp: "))

# Lặp từ 1 đến n
for i in range(1, n + 1):
    print("Lần lặp for:", i)
```

```python
# 4.b. Lệnh lặp While

# Nhập một số từ người dùng
limit = int(input("Nhập số lần lặp: "))

count = 1  # Giá trị bắt đầu đếm

# Lặp cho đến khi count lớn hơn limit thì dừng
while count <= limit:
    print("lần lặp while:", count)  # In giá trị đang đếm
    count += 1  # Tăng biến đếm thêm 1 mỗi vòng lặp
```

```python
#HÀM CÓ GIÁ TRỊ TRẢ VỀ 
# Hàm tính tổng hai số và trả về kết quả.

def tinh_tong(a, b):
    tong = a + b
    return tong  # Trả về giá trị cho người gọi

# Nhập 2 số
x = float(input("Nhập số thứ nhất: "))
y = float(input("Nhập số thứ hai: "))

# Gọi hàm và in kết quả
ket_qua = tinh_tong(x, y)
print("Tổng hai số là:", ket_qua)
```

6.Đọc ghi file văn bản

```python
#GHI FILE VĂN BẢN
# Mục tiêu ghi một đoạn văn bản vào file "output.txt"

# Mở file ở chế độ ghi ('w')
# Nếu file chưa tồn tại -> Python tự tạo mới
with open("output.txt", "w", encoding="utf-8") as f:
    f.write("MSSV: 082205010652\n")
    f.write("Họ tên: Lê Minh Liêm\n")

print("Đã ghi dữ liệu vào file output.txt")
```

```python
#ĐỌC TOÀN BỘ FILE
# Mục tiêu đọc lại toàn bộ nội dung vừa ghi

with open("output.txt", "r", encoding="utf-8") as f:
    content = f.read()  # Đọc toàn bộ nội dung file

print("Nội dung file:")
print(content)
```

```python
#ĐỌC FILE THEO TỪNG DÒNG 
# Mục tiêu đọc và xử lý từng dòng một

with open("output.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()  # Trả về danh sách các dòng

print("Các dòng trong file:")
for line in lines:
    print(line.strip())  # strip() để bỏ xuống dòng dư thừa
```
