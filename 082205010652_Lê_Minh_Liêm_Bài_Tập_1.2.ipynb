1.  Đếm xem có bao nhiêu dòng và bao nhiêu dòng xuất hiện spam và bao nhiêu dòng xuất hiện ham trong file trên

```python
num_lines = 0     
num_spam = 0      
num_ham = 0      

with open("SMSSpamCollection") as f:  
    for line in f:                    
        num_lines += 1                

        # Kiểm tra nhãn đầu dòng
        if line.startswith("spam"):
            num_spam += 1
        elif line.startswith("ham"):
            num_ham += 1

print("Tổng số dòng:", num_lines)
print("Số dòng spam:", num_spam)
print("Số dòng ham:", num_ham)
```

2. Đếm xem có bao nhiêu từ (các từ ngăn cách bởi khoảng trắng, và không phải chữ số)

```python
word_count = 0     

with open("SMSSpamCollection") as f:
    for line in f:
        words = line.split()          # tách từ bằng khoảng trắng
        for w in words:
            # numeric = toàn chữ số không đếm
            if not w.isnumeric():
                word_count += 1

print("Tổng số từ (không tính từ toàn chữ số):", word_count)
```

3. Đếm xem có bao nhiêu từ 'you' (không phân biệt hoa thường)

```python
you_count = 0

with open("SMSSpamCollection") as f:
    for line in f:
        # chuyển cả câu về lower để so sánh chính xác
        words = line.lower().split()

        # đếm từng từ
        for w in words:
            if w == "you":
                you_count += 1

print("Số lần xuất hiện từ 'you':", you_count)
```

4. Đếm xem có bao nhiêu từ khác nhau (không phân biết hoa thường)

```python
unique_words = set()   # tập chứa các từ không trùng

with open("SMSSpamCollection") as f:
    for line in f:
        words = line.lower().split()
        for w in words:
            if not w.isnumeric():      # bỏ từ toàn số
                unique_words.add(w)    # thêm vào set

print("Số lượng từ khác nhau:", len(unique_words))
```

5. Liệt kê 10 từ khóa xuất hiện nhiều nhất trong file dữ liệu (không phân biết hoa thường)

```python
from collections import Counter

counter = Counter()   # tạo bộ đếm

with open("SMSSpamCollection") as f:
    for line in f:
        words = line.lower().split()
        for w in words:
            if not w.isnumeric():      # bỏ từ toàn số
                counter[w] += 1        # tăng số lần xuất hiện

# lấy 10 từ có tần suất cao nhất
top10 = counter.most_common(10)

print("Top 10 từ xuất hiện nhiều nhất:")
for word, freq in top10:
    print(f"{word}: {freq}")
```
